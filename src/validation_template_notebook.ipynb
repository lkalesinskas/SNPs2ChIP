### Example: Asthma SNP 17:38023745

Identification of the key component and show the GREAT results for the top component.
We have GREAT results for the following ontologies:

- `GOBiologicalProcess`: Gene ontology biological process.
- `GOCellularComponent`: Gene ontology cellular component.
- `GOMolecularFunction`: Gene ontology molecular function.
- `MGIPhenotype`: Mouse genome informatics (MGI) phenotype ontology. Compared to human, mouse data has relatively comprehensive catalog of gene-phenotype mapping.
- `MGIPhenoSingleKO`: This is a subset of MGI phenotype. The gene-phenotype annotations via gene knock out (KO) experiments are included in this subset.
- `HumanPhenotypeOntology`: Human phenotype ontology.

As an illustration, we just report the results for `HumanPhenotypeOntology` and `GOBiologicalProcess`.

```python
decomposed_mats.find_pcs_given_locus(
    genomic_bins.find_locus_given_snp(17, 38023745)
)
```

```python
read_great_res_wrapper(data_dir, 7, 'HumanPhenotypeOntology', topk=10)
```

```python
read_great_res_wrapper(data_dir, 7, 'GOBiologicalProcess', topk=10)
```
